```python
import numpy
import pandas
import matplotlib.pyplot as plt
#기계공학부 20191089 김찬우 기계공학실험설계 1조 실험 2주차 유체역학 실험. 파이썬 그래프 도시 프로그래밍.
```

```python
data = pandas.read_csv('exp2 data.csv')
print(data)
#반경에 따른 속도 분포와 오차를 불러옴
```

```python
plt.figure(figsize=[15,4])

plt.subplot(1,3,1)
plt.errorbar(data.r[0:5], data.v[0:5], yerr=data.v[0:5]*(data.err[0:5]/100),  fmt="bo", elinewidth=0.5, capsize=2)
plt.xticks(numpy.arange(1.5,15,3))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 15Hz")

plt.subplot(1,3,2)
plt.errorbar(data.r[5:10], data.v[5:10], yerr=data.v[5:10]*(data.err[5:10]/100),  fmt="bo", elinewidth=0.5, capsize=2)
plt.xticks(numpy.arange(1.5,15,3))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 30Hz")

plt.subplot(1,3,3)
plt.errorbar(data.r[10:15], data.v[10:15], yerr=data.v[10:15]*(data.err[10:15]/100),  fmt="bo", elinewidth=0.5, capsize=2)
plt.xticks(numpy.arange(1.5,15,3))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 45Hz")

plt.show()
#반경에 따른 속도를 그래프에 표시하고 에러바 도시.
```

```python
plt.figure(figsize=[15,4])

plt.subplot(1,3,1)
a = 7.882414
R = 15
x = numpy.arange(0,15,0.01)
y = a * (1 - (x/R)) ** (1/7)
plt.plot(x, y, "--", color="gray")
plt.plot(data.r[0:5], data.v[0:5], "s",  color="black")
plt.xticks(numpy.arange(0,15.1,1.5))
plt.yticks(numpy.arange(3,8.1,1))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 15Hz")
plt.text(3, 3.5, f"v = {a}( 1 - (r/{R})^(1/7) )")

plt.subplot(1,3,2)
a = 16.59719
R = 15
x = numpy.arange(0,15,0.01)
y = a * (1 - (x/R)) ** (1/7)
plt.plot(x, y, "--", color="gray")
plt.plot(data.r[5:10], data.v[5:10], "s",  color="black")
plt.xticks(numpy.arange(0,15.1,1.5))
plt.yticks(numpy.arange(5,17.1,1))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 30Hz")
plt.text(3, 7, f"v = {a}( 1 - (r/{R})^(1/7) )")

plt.subplot(1,3,3)
a = 24.72811
R = 15
x = numpy.arange(0,15,0.01)
y = a * (1 - (x/R)) ** (1/7)
plt.plot(x, y, "--", color="gray")
plt.plot(data.r[10:15], data.v[10:15], "s",  color="black")
plt.ylim(12,25.1)
plt.xticks(numpy.arange(0,15.1,1.5))
plt.yticks(numpy.arange(12,25.1,1))
plt.xlabel("radius (mm)")
plt.ylabel("velocity (m/s)")
plt.title("Blower frequency = 45Hz")
plt.text(3, 14, f"v = {a}( 1 - (r/{R})^(1/7) )")

plt.show()
#1/7제곱법칙을 이용해 난류속도분포 식 맞춤 및 도시.
```
